```python
import pandas as pd
import numpy as np
import wordcloud as wc
from matplotlib import pyplot as plt

data = pd.read_excel('data.xlsx', sheet_name = 'Lab 1 - Web Intelligence Public')
data.head(5)
```

```python
data_2000_2006 = data[data['Year'].between(2000 ,2006 ,inclusive = True)]
```

```python
keywords_list_raw = data_2000_2006['Author Keywords'].dropna().tolist()
keyword_list = []
for keywlist in keywords_list_raw:
    for word in keywlist.split(';'):
        keyword_list.append(word.lower().strip())
len(keyword_list)
```

```python
frequencies = pd.Series(keyword_list).value_counts()
frequencies
```

```python
wordcloud = wc.WordCloud().generate(''.join(keyword_list))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
