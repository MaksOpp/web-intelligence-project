```python
import pandas as pd
import numpy as np
import wordcloud as wc
from matplotlib import pyplot as plt

data = pd.read_excel('data.xlsx', sheet_name = 'Lab 1 - Web Intelligence Public')
data.head(5)
```

```python
data_2000_2006 = data[data['Year'].between(2000 ,2006 ,inclusive = True)]
```

```python
# author keywords wordcloud
author_keywords_list_raw = data_2000_2006['Author Keywords'].dropna().tolist()
author_keyword_list = []
for keywlist in author_keywords_list_raw:
    for word in keywlist.split(';'):
        author_keyword_list.append(word.lower().strip())
author_frequencies = pd.Series(author_keyword_list).value_counts()
auth_wordcloud = wc.WordCloud().generate_from_frequencies(author_frequencies)
plt.imshow(auth_wordcloud)
plt.axis("off")
plt.show()
```

```python
# index keywords wordcloud
index_keywords_list_raw = data_2000_2006['Index Keywords'].dropna().tolist()
index_keyword_list = []
for keywlist in index_keywords_list_raw:
    for word in keywlist.split(';'):
        index_keyword_list.append(word.lower().strip())
index_frequencies = pd.Series(index_keyword_list).value_counts()
index_wordcloud = wc.WordCloud().generate_from_frequencies(index_frequencies)
plt.imshow(index_wordcloud)
plt.axis("off")
plt.show()
```
